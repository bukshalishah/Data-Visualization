```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame([1,2,3,57,2,424,623,234], [34,25,74,234,53,234,63,23])
```

```python
df
```

# Histrogram

```python
sns.distplot(df, kde=True)
```

```python
sns.distplot(df, kde=False)
```

```python
# Create a distplot of the Award Amount
sns.distplot(df,
             hist=False,
             rug=True,
             kde_kws={'shade':True})

# Plot the results
plt.show()
```
